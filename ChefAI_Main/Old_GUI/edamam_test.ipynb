```python
import requests
from pprint import pprint
from APIkeys import ApplicationID, ApplicationKey   # Aden, change this to your own API key file
```

```python
# configure request
URL = "https://api.edamam.com/api/recipes/v2"
query = "chicken, olives, tomatoes, garlic, onion, pepper" # pretend these are ingredients

# https://developer.edamam.com/edamam-docs-recipe-api
params = {"q": query, "app_id": ApplicationID, "app_key": ApplicationKey, 
                        "type":"any", # search for any type of recipe
                        "ingr":"6-10", # number of ingredients, maybe with some more then what we use in query
                        "from":0, "to":3, # return the first 3 recipes
                        "time": "0-20", # time in minutes, 0-20 minutes
                        "diet": "balanced", # return only balanced recipes
                        "health":"peanut-free",
                        "calories":"500-900",
                        "excluded":"bell peppers",
                        "cuisineType":["American", "European"], # a list means either American or European
                        "dishType":"main course",
                        "mealType":["dinner", "lunch"],
                        # return only these fields in the JSON results
                        # label is the title of the recipe
                        "field":["label","image", "calories","ingredientLines","yield", "totalTime", "url"],
                        # uri, label, image, images, source, url, shareAs, yield, dietLabels, healthLabels, cautions, ingredientLines, ingredients, calories, glycemicIndex, inflammatoryIndex, totalCO2Emissions, co2EmissionsClass, totalWeight, totalTime, cuisineType, mealType, dishType, totalNutrients, totalDaily, digest, tags, externalId
                        # url is the url of the website with the recipe, so you always just use that to show your user
                        # instructions the recipe
            }
```

```python
response = requests.get(URL, params=params)
print(response.status_code)
#print(response.url)

result = response.json()
recipes = result['hits']
print("got", len(recipes), "hits")
```

```python
from IPython.display import Image
r = recipes[0]['recipe']
print(r['label'])
image_url = r['image']
display(Image(url=image_url))
```

```python
# show all values for first recipe
pprint(r)
```

```python
# sort by calories
sorted_recipes = sorted(recipes, key=lambda x: x['recipe']['calories'])
for r in sorted_recipes:
    print(r['recipe']['label'], r['recipe']['calories'])
```
